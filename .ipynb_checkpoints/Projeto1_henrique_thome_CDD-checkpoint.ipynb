### Pergunta: A democracia tEm relação com o IDH? E o índice de suicídio, têm relação com o IDH?

### Descrição: 
    A democracia é o sistema mais comúm atualmente e existe uma forte ideia de que esse está diretamente atrelado a uma melhor qualidade de vida da sociedade. Exemplos como Cuba e a China, por exemplo, me fazem refletir sobre essa afirmação, que já virou parte do senso comúm, portanto, decidi investigar se essa é fundamentada em fatos reais.
    O IDH mede a qualidade de vida dos países, assim, acredita-se que quanto maior esse índice mais a população é feliz. A felicidade é algo difícil de ser mensurado, por tanto para ver se essa correlação é verdade, decidi verificar se o IDH está relacionado com o índice de suicídio a cada dez mil habitantes. Essa variavel foi escolhida, pois têm uma relação direta com a felicidade.

```python
import pandas as pd
import numpy as np
import math as math
from scipy import stats
import matplotlib.pyplot as plt
```

```python

IDH = pd.read_excel('hdi_human_development_index.xlsx')
IDH = IDH.rename(index=str, columns={"geo": "IDH"})
democracy = pd.read_excel('democracy_score_use_as_color.xlsx')
democracy = democracy.rename(index=str, columns={"geo": "Democracy"})
suicide = pd.read_excel('suicide_per_100000_people.xlsx')
suicide = suicide.rename(index=str, columns={"geo": "suicide/100000"})
```

```python
IDH_2005 = IDH.loc[:, ["IDH",2005]]
democracy_2005 = democracy.loc[:, ["Democracy",2005]]
suicide_2005 = suicide.loc[:, ["suicide/100000",2005]]
```

```python
IDH_idx = IDH_2005.set_index('IDH')
```

```python
democracy_idx = democracy_2005.set_index('Democracy')
```

```python
suicide_idx = suicide_2005.set_index('suicide/100000')
```

```python
data_1 = IDH_idx.join(democracy_idx, how='inner', lsuffix='_IDH', rsuffix='_democracy')
data = data_1.join(suicide_idx, how="inner", lsuffix='_suicide')
data["suicide per 100000"] = data.pop(2005)
data.dropna()
```

```python
def dispersao_com_reta(dados, variavel_x, variavel_y):
    dados_X = dados[variavel_x]
    dados_Y = dados[variavel_y]
    
    # Calcula os coeficientes da reta de ajuste.
    cov = dados_X.cov(dados_Y)

    var_X = dados_X.var()
    var_Y = dados_Y.var()

    mX = dados_X.mean()
    mY = dados_Y.mean()
    
    a = cov / var_X
    b = mY - a*mX

    # Desenha o gráfico de dispersão e a reta de ajuste.
    min_X = dados_X.min()
    max_X = dados_X.max()

    plt.scatter(dados_X, dados_Y)  # dados.plot.scatter(x=variavel_X, y=variavel_Y) tambem serve!
    plt.plot((min_X, max_X), (a*min_X + b, a*max_X + b), color='red')

    plt.title('{0} versus {1}'.format(variavel_x, variavel_y))
    plt.xlabel(variavel_x)
    plt.ylabel(variavel_y)
    
    # Retorna os coeficientes de ajuste da reta.
    return a, b
```

#### Verificando a relação entre IDH e democracia

```python

a, b = dispersao_com_reta(data, variavel_x="2005_IDH", variavel_y="2005_democracy")
print('Coeficiente linear:', b)
print('Coeficiente angular:', a)
```

```python
data.loc[:,["2005_IDH", "2005_democracy"]].corr()
```

#### Verificando a relação entre IDH e suicídio por 10000 habitante

```python
a, b = dispersao_com_reta(data, variavel_x="2005_IDH", variavel_y='suicide per 100000')
print('Coeficiente linear:', b)
print('Coeficiente angular:', a)
```

```python
data.loc[:,["2005_IDH", 'suicide per 100000']].corr()
```

### Conclusão


##### IDH vs Democracia:
Observando o gráfico não é possível estabelecer ao certo se existe relação entre as duas variáveis. Por um lado, a reta de ajuste resultou em um coeficiente angular de 13, o que mostra que a associação é positiva, corroborando para a tese de que quanto mais democrático maior o IDH. Por outro lado o coeficiente de associação de Pearson foi calculado, de modo a determinar a força dessa relação. O resultado foi de r=0.35, o que indica que há uma baixissima relação entre os dois. Dessa forma, não é coerente dizer que a democracia esta relacionada com o IDH, o que derruba minha hipótese a o senso comúm.

##### IDH vs Suicídio por 100000 habitantes
Observando o gráfico não é possível estabeler uma relação entre IDH e suicídio. O coeficiente de Pearson é de r = 0.06, o que mostra uma correlação virtualmente nula. A partir disso é possível perceber que, apesar do IDH estar diretamente relacionado com uma melhor qualidade de vida a felicidade das pessoas não está atrelado a isso.
